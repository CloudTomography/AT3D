```python
@xr.register_dataarray_accessor("GridData")
class GridData(object):
    def __init__(self, data, coords, dims):
        self._obj = xr.DataArray(data=data,coords=coords,dims=dims)
        self._nx = None
    
    @property
    def nx(self):
        if self._nx is None:
            self._nx = len(self._obj.coords['x'])
        return self._nx

    
```
